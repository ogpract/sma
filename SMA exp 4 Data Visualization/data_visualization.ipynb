```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np 
from sklearn.metrics import confusion_matrix 
%matplotlib inline 

df = pd.read_csv('tik_tok_analytics.csv') 
df = df.head(10) 
print("First 10 rows of the dataset:") 
print(df)
```

```python
numeric_columns = ['Views', 'Shares', 'Comments', 'Likes', 'Length', 'avg_watch'] 
existing_numeric_columns = [col for col in numeric_columns if col in df.columns] 

if existing_numeric_columns: 
    plt.figure(figsize=(10, 6)) 
    sns.boxplot(data=df[existing_numeric_columns]) 
    plt.title('Box Plot for Numeric Columns (10 Samples)') 
    plt.show() 
else: 
    print("None of the specified numeric columns found for the box plot.")
```

```python
if 'Views' in df.columns: 
    plt.figure(figsize=(8, 5)) 
    plt.hist(df['Views'].dropna(), bins=10, edgecolor='k', alpha=0.7) 
    plt.title('Histogram of Views (10 Samples)') 
    plt.xlabel('Views') 
    plt.ylabel('Frequency') 
    plt.show() 
else: 
    print("Column 'Views' not found for the histogram.")
```

```python
if 'Views' in df.columns: 
    plt.figure(figsize=(10, 6)) 
    plt.plot(df.index, df['Views'], marker='o', linestyle='-') 
    plt.title('Line Chart of Views Over Index (10 Samples)') 
    plt.xlabel('Index') 
    plt.ylabel('Views') 
    plt.show() 
else: 
    print("Column 'Views' not found for the line chart.")
```

```python
if 'Day' in df.columns: 
    plt.figure(figsize=(8, 8)) 
    day_counts = df['Day'].value_counts() 
    day_counts.plot.pie(autopct='%1.1f%%', startangle=90) 
    plt.title('Pie Chart of Day (10 Samples)') 
    plt.ylabel('')  # Hide the default y-label 
    plt.show() 
else: 
    print("Column 'Day' not found for the pie chart.")
```

```python
if 'Views' in df.columns and 'Likes' in df.columns: 
    plt.figure(figsize=(8, 6)) 
    sns.scatterplot(x='Views', y='Likes', data=df) 
    plt.title('Scatter Plot: Views vs Likes (10 Samples)') 
    plt.xlabel('Views') 
    plt.ylabel('Likes') 
    plt.show() 
else: 
    print("Columns 'Views' and/or 'Likes' not found for the scatter plot.")
```
